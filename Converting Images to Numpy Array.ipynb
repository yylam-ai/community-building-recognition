```
import os 
tmp_train_path_sorted = sorted(os.listdir("/content/drive/MyDrive/Colab Notebooks/Building Datasets/train_dataset"))
```

```
train_path_sorted = []

for i in range(len(tmp_train_path_sorted)):
    train_path_sorted.append("train_dataset/"+tmp_train_path_sorted[i])
    
```

```
X_train = np.array([])
y_train = np.array([])
count = 0

for image in train_path_sorted:
    with open(image, 'rb') as file:
        img = cv2.imread(image, 0)
        count += 1
        reshaped_img = np.reshape(img, img.shape[0]*img.shape[1])
        transposed_img = np.array([reshaped_img]).T
        X_train = np.append(X_train, transposed_img)
        char = image[-7]
        y_train = np.append(y_train, char)
        
X_train = np.reshape(X_train, (count, reshaped_img.shape[0]))
y_train = np.uint8(y_train)  
    
X_train.shape
```

```
tmp_test_path_sorted = sorted(os.listdir("test_dataset"))
```

```
test_path_sorted = []

for i in range(len(tmp_test_path_sorted)):
    test_path_sorted.append("test_dataset/"+tmp_test_path_sorted[i])
    
```

```
test_path_sorted
```

```
X_test = np.array([])
y_test = np.array([])
count = 0

for image in test_path_sorted:
    with open(image, 'rb') as file:
        img = cv2.imread(image, 0)
        count += 1
        reshaped_img = np.reshape(img, img.shape[0]*img.shape[1])
        transposed_img = np.array([reshaped_img]).T
        X_test = np.append(X_test, transposed_img)
        char = image[-7]
        y_test = np.append(y_test, char)
        
X_test = np.reshape(X_test, (count, reshaped_img.shape[0]))      
y_test = np.uint8(y_test)
    
X_test.shape
```

```
y_test
```

```
hf = h5py.File('dataset.h5', 'w')
```

```
hf.create_dataset('X_test', data=X_test)
hf.create_dataset('y_test', data=y_test)
hf.create_dataset('X_train', data=X_train)
hf.create_dataset('y_train', data=y_train)
```

```
hf.close()
```
